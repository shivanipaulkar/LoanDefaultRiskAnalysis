Simple Interpretation for Your Project
Older, higher-income, and longer-employed borrowers tend to default less.
Higher interest rates and larger loan amounts slightly increase default risk.
Most variables show very weak correlations, meaning default likely depends on multiple combined factors rather than a single variable.

“The correlation analysis shows that Age has the strongest negative relationship with default (-0.17), meaning older borrowers tend to default less. Interest rate has the strongest positive correlation (0.13), indicating that higher interest rates increase the probability of default. However, most variables show weak correlations, suggesting that **loan default is influenced by multiple factors rather than a single variable.”

```python
## Income Group
df['IncomeGroup'] = pd.cut(
    df['Income'],
    bins=[0,30000,50000,100000,130000,150000],
    labels=['Low','Lower-Mid','Middle','Upper-Mid','High']
)

df['IncomeGroup'].value_counts()
```

```python
## Credit Score Group
df['CreditScoreGroup']=pd.cut(
    df['CreditScore'],
        bins=[200,580,670,740,800,850],
        labels=['Poor','Fair','Good','Very Good','Excellent']
)

df['CreditScoreGroup'].value_counts()
```

```python
## Age Group
df['AgeGroup']=pd.cut(
    df['Age'],
    bins=[15,30,40,50,60,70],
    labels=['18-30','31-40','41-50','51-60','60+']
)

df['AgeGroup'].value_counts()
```

```python
## Loan Size Group
df['LoanSize']=pd.cut(
    df['LoanAmount'],
    bins=[0,50000,100000,150000,200000,250000],
    labels=['Very Small', 'Small', 'Medium', 'Large', 'Very Large']
)

df['LoanSize'].value_counts()
```

```python
## Converting Default Column to yes no format 
df['DefaultStatus']=df['Default'].map({0:'No', 1:'Yes'})
```
