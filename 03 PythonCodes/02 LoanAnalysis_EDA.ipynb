```python
## Imported the required libraries.
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## reading a csv saved by performing feature engineering.
df=pd.read_csv('LoanPythonCleanedData.csv')
```

```python
df.head()
df.columns
```

Business Questions to answer (Using crosstab)

1️⃣ Which income group has the highest loan default rate?
   Low-income borrowers show the highest default rate (~22%), while high-income borrowers have the lowest.


2️⃣ Does credit score affect loan default risk?
   Yes, borrowers with lower credit scores have a higher probability of default. So credit Score is an important factor.


3️⃣ Which age group shows the highest default probability?
   Young borrowers (18–30) have the highest default rate, while older borrowers default less.


4️⃣ Which employment type is most risky for loan default?
   Unemployed borrowers show the highest default rate, while full-time employees have the lowest.


5️⃣ Does loan size influence the probability of default?
   Larger loans have higher default rates, while smaller loans are safer.

```python
## 1. Which Income Group has the highest default rate?
pd.crosstab(df['IncomeGroup'], df['DefaultStatus'], normalize='index')*100
```

```python
## Income vs Default
pd.crosstab(df['IncomeGroup'], df['DefaultStatus'], normalize='index').plot(kind='bar', stacked=True, figsize=(5,2), color=['skyblue','pink'])
plt.title("Loan Default Rate by Income Group")
plt.ylabel('Percentage')
plt.legend(title='DefaultStatus', loc='lower right')
plt.show()
```

```python
## 2. Does Credit Score affect loan default?
pd.crosstab(df['CreditScoreGroup'], df['DefaultStatus'], normalize='index')*100
```

```python
## Credit Score vs Defaults
pd.crosstab(df['CreditScoreGroup'], df['DefaultStatus'], normalize='index').plot(kind='bar', stacked=True, figsize=(5,2), color=['#CDB4DB', '#FFD6A5'])
plt.title('Loan Default Rate by Credit Score')
plt.legend(title='Default Status', loc='lower right')
plt.show()
```

```python
## 3. Which Age Group defaults more?
pd.crosstab(df['AgeGroup'], df['DefaultStatus'], normalize='index')*100
```

```python
## Age vs Default
pd.crosstab(df['AgeGroup'], df['DefaultStatus'], normalize='index').plot(kind='bar', stacked=True, figsize=(5,2), color=['skyblue', 'pink'])
plt.title("Loan Default Rate by Age Group")
plt.legend(title='Default Status', loc='lower right')
plt.show()
```

```python
## 4. Which Employment Type is Riskier?
pd.crosstab(df['EmploymentType'], df['DefaultStatus'], normalize='index')*100
```

```python
## Employement vs Default
pd.crosstab(df['EmploymentType'], df['DefaultStatus'], normalize='index').plot(kind='bar', stacked=True, figsize=(5,2), color=['#CDB4DB', '#FFD6A5'])
plt.title("Loan Default Rate by Employment Type")
plt.legend(title='Default Status', loc='lower right')
plt.show()
```

```python
## 5. Does Loan Size affect Default?
pd.crosstab(df['LoanSize'], df['DefaultStatus'], normalize='index')*100
```

```python
pd.crosstab(df['LoanSize'], df['DefaultStatus'], normalize='index').plot(kind='bar', stacked=True, figsize=(5,2), color=['skyblue', 'pink'])
plt.title("Loan Default Rate by Loan Size")
plt.legend(title='Default Status', loc='lower right')
plt.show()
```
